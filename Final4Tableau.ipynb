## GET ALL THE SESSION OBJECTS FROM THE API

## GET THE FASTEST LAP TELEMETRY DATA FOR ALL DRIVERS FOR ALL SESSIONS OF THE ITALIAN GRAND PRIX 

```python
for session in sessions[14:15]:
    gp_path = os.path.join(root_path,session.name)
    print(session.name)
    try:
        practice1_laps = session.get_practice(1).load_laps(with_telemetry = True) # pick practice event in sessions
        practice2_laps = session.get_practice(2).load_laps(with_telemetry = True) # pick practice event in sessions
        practice3_laps = session.get_practice(3).load_laps(with_telemetry = True) # pick practice event in sessions
        qualifying_laps = session.get_quali().load_laps(with_telemetry = True) # pick practice event in sessions
#     if driver=="MAZ":
#         continue
    except UnboundLocalError:
        pass
    except KeyError:
        pass
    
    for driver in drivers:                                       #   pick driver
        fastest_lap_telemetry_driver_path = os.path.join(gp_path, "_Fastest_Laps_Telemetry")
        print(fastest_lap_telemetry_driver_path)
        if not os.path.exists(fastest_lap_telemetry_driver_path):
            os.mkdir(fastest_lap_telemetry_driver_path)
        
        car_data_df = pd.DataFrame()
        telementry_data_df = pd.DataFrame()
        position_data_df = pd.DataFrame()
        weather_data_df = pd.DataFrame()
#         fastest_lap_df = pd.DataFrame()
#         print("Driver out:", driver)
        try:
            car_fastest_lap_telemetry_driver_practice1 = practice1_laps.pick_driver(driver).pick_fastest()
            car_fastest_lap_telemetry_driver_practice1 = car_fastest_lap_telemetry_driver_practice1.get_telemetry()

            car_fastest_lap_telemetry_driver_practice2 = practice2_laps.pick_driver(driver).pick_fastest()
            car_fastest_lap_telemetry_driver_practice2 = car_fastest_lap_telemetry_driver_practice2.get_telemetry()

            car_fastest_lap_telemetry_driver_practice3 = practice3_laps.pick_driver(driver).pick_fastest()
            car_fastest_lap_telemetry_driver_practice3 = car_fastest_lap_telemetry_driver_practice3.get_telemetry()

            car_fastest_lap_telemetry_driver_qualifying = qualifying_laps.pick_driver(driver).pick_fastest()
            car_fastest_lap_telemetry_driver_qualifying = car_fastest_lap_telemetry_driver_qualifying.get_telemetry()

            savefile1 = '%s_%s_%s_%s%s'%(session.name,events_filenames[0],"fastest_lap_telemetry_data",driver,'.csv')
            savefile2 = '%s_%s_%s_%s%s'%(session.name,events_filenames[1],"fastest_lap_telemetry_data",driver,'.csv')
            savefile3 = '%s_%s_%s_%s%s'%(session.name,events_filenames[2],"fastest_lap_telemetry_data",driver,'.csv')
            savefile4 = '%s_%s_%s_%s%s'%(session.name,events_filenames[3],"fastest_lap_telemetry_data",driver,'.csv')
#             print("Driver In:", driver)
            car_fastest_lap_telemetry_driver_practice1_df = pd.DataFrame(car_fastest_lap_telemetry_driver_practice1)
            car_fastest_lap_telemetry_driver_practice2_df = pd.DataFrame(car_fastest_lap_telemetry_driver_practice2)
            car_fastest_lap_telemetry_driver_practice3_df = pd.DataFrame(car_fastest_lap_telemetry_driver_practice3)
            car_fastest_lap_telemetry_driver_qualifying_df = pd.DataFrame(car_fastest_lap_telemetry_driver_qualifying)
            
            car_fastest_lap_telemetry_driver_practice1_df["Driver1"] = driver
            car_fastest_lap_telemetry_driver_practice2_df["Driver1"] = driver
            car_fastest_lap_telemetry_driver_practice3_df["Driver1"] = driver
            car_fastest_lap_telemetry_driver_qualifying_df["Driver1"] = driver
            
            
            
            car_fastest_lap_telemetry_driver_practice1_df.to_csv(os.path.join(fastest_lap_telemetry_driver_path,savefile1),sep = ";")
            car_fastest_lap_telemetry_driver_practice2_df.to_csv(os.path.join(fastest_lap_telemetry_driver_path,savefile2),sep = ";")
            car_fastest_lap_telemetry_driver_practice3_df.to_csv(os.path.join(fastest_lap_telemetry_driver_path,savefile3),sep = ";")
            car_fastest_lap_telemetry_driver_qualifying_df.to_csv(os.path.join(fastest_lap_telemetry_driver_path,savefile4),sep = ";")
        
        except ValueError:
            print("Value Error in driver: ", driver)
            pass
```
